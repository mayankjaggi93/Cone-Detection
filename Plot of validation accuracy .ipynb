```python
import pandas as pd
df=pd.read_excel('val_accuracy.xlsx')
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(15,10), dpi=100)

axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

axes.plot(df['Epoch'], df['Validation Accuracy'], 'b')

axes.set_xlabel('Epoch')
axes.set_ylabel('Validation Accuracy ')
axes.set_title('Plot Tracking Validation Accuracy across Epoch')

z = np.polyfit(df['Epoch'], df['Validation Accuracy'], 1)
p = np.poly1d(z)
plt.plot(df['Epoch'],p(df['Epoch']),"r--")
```
